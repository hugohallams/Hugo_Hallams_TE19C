```python
#Uppgift 1 g

import matplotlib.pyplot as plt
import random
import math

cirkel = plt.Circle((0,0), radius=1, color="black")#Skapar radien och färg på cirkeln

fig, ax = plt.subplots()
ax.add_artist(cirkel)

for punkter in range(1000):   #simulerar en x och en y kordinat 1000 gånger
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    

    if x**2 + y**2 <  1**2:
        plt.plot(x,y,".g") #Gör alla punkter inne i cirkeln gröna

    else:
        plt.plot(x,y,".r") #Gör alla punkter utanför cirkeln röda


plt.show()
```

```python
#2 d

burr = int(input("Vilket tal vars multipel ska bli burr?"))
birr = int(input("Vilket tal vars multipel ska bli birr?"))

for tal in range(0,101):
    if tal%birr == 0 and tal%burr == 0: #För att göra så att både burr och birr skrivs ur så anväder jag mig av "and" som gör att om båda kraven uppfylls skrivs burr,birr ut.
        print("burr, birr")

    elif tal%burr == 0:
        print("burr")

    elif tal%birr == 0:
        print("birr")

    
    else:
        print(tal)
```

```python
#2 e

start = int(input("Vart ska leken börja?"))# man måste göra inputen till int så att den alltid ses som ett tal
slut = int(input("Vart ska leken sluta?"))
burr = int(input("Vilket tal vars multipel ska bli burr?"))
birr = int(input("Vilket tal vars multipel ska li birr?"))
antalburr = 0
antalbirr = 0
antalbirrburr = 0

for tal in range(start,slut + 1):
    if tal%birr == 0 and tal%burr == 0:
        print("burr, birr")
        antalbirrburr += 1 # gör så att för varje burrbirr läggs det till 1 på variabeln

    elif tal%burr == 0:
        print("burr") 
        antalburr += 1 #Gör så att för varje burr läggs det till 1 på variabeln.

    elif tal%birr == 0:
        print("birr")
        antalbirr += 1 #Gör så att för varje birr läggs det till 1 på variabeln.

    else:
        print(tal)

print(f"Antal burr är {antalburr}. Antal birr är {antalbirr}. Antal birr,burr är {antalbirrburr}.")
```
