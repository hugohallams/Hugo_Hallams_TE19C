```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,10)

print(f"x = {x}")

# y = kx+m

def f(k,x,m):
    y = k*x + m
    return y 

#y = x+1
print(f" y = {f(1,x,1)}")



plt.plot(x,f(1,x,1))
plt.title("Rät linje")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
